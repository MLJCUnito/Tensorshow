```
# So, We need to convert temperature to SI
def FtoC(F):
  return np.round((F - 32) * 5 /9,1)

#How can we do this?
```

```
# we use .values to convert pandas array which are coalled Series to numpy array
FtoC(data.TAVG.values)
```

```
#or use the apply function
data.TAVG.apply(FtoC)
```

```
# set the databases values to the new ones
data.TAVG = data.TAVG.apply(FtoC)
data.TMAX = data.TMAX.apply(FtoC)
data.TMIN = data.TMIN.apply(FtoC)
data.head()
```

```
# convert them to datetime format
data.DATE = pd.to_datetime(data.DATE,format='%Y-%m-%d')
data.head()
```

# Data Analysis

```
# Missing values in %
data.count()/len(data)*100
```

```
import matplotlib.dates as mdates

F = plt.figure(figsize=(32,5))
plt.plot(data.DATE,data.TAVG)
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator()) #ticks on 1st january
plt.grid()
```

```
# groupby function: we want for example group our dataset by years

# we don't wanna mess around with the original dataset so we create a copy
df = data.copy() 

# so groupby function let you do function to single class, for example say we want
# to take the mean of the temperature by year
df = df.groupby(df.DATE.dt.year)
df = df.mean()
df.tail()
```

```
#NB we plot the index as x because we grouped our dataset by DATE
plt.plot(df.index,df.TAVG)
```

Strange variation appears... uhmmmmmmm, let's see how many data we have per year

```
#Always create a copy
df = data.copy()

#Add a column with all one to sum up and check the number of data per year
# to add columns we have to use dictionary syntax
df['COUNTS'] = 1
df = df.groupby(df.DATE.dt.year).sum()
plt.plot(df.index,df.COUNTS)
```

Not a uniform distribution isn't it? So we have to delete first and last years to have a fair comparison

```
#how we can select a specific year?
data.DATE.dt.year==1973
```

```
#now we can use this boolean Series to Select part of the dataframe
data[data.DATE.dt.year==1973]
```

```
#or the not one
data[data.DATE.dt.year!=1973]
```

```
#delete all the date we don't want
data = data[data.DATE.dt.year!=1973]
data = data[data.DATE.dt.year!=1974]
data = data[data.DATE.dt.year!=2019]
```

now plot the moving average on top of the series

```
df = data.copy()
df = df.groupby(df.DATE.dt.year).mean()

#the rolling mean function is like the groupby one, first we apply rolling to the
# dataframe then a specific function like the mean
rolling_mean = df.TAVG.rolling(window=10).mean()

plt.plot(df.index,df.TAVG,'-')
plt.plot(df.index,rolling_mean)

plt.grid()
```

Another useful information is the quantile 

https://en.wikipedia.org/wiki/Quantile

```
# obviously there's a pandas function already implemented
df.quantile(0.75).TAVG
```

```
df = data.copy()
df = df.groupby(df.DATE.dt.year)

plt.plot(df.max().TAVG.rolling(window=10).mean(),label='max')

plt.plot(df.quantile(0.75).TAVG.rolling(window=10).mean(),label='75%')

plt.plot(df.quantile(0.5).TAVG.rolling(window=10).mean(),label='50%')

plt.plot(df.quantile(0.25).TAVG.rolling(window=10).mean(),label='25%')

plt.plot(df.min().TAVG.rolling(window=10).mean(),label='min')

plt.legend()

plt.grid()
```

```
import plotly.express as px
```

```
fig = px.scatter(df, 
                 x='DATE',
                 y='TAVG',
                 template="plotly_dark")
fig.show()
```

```
fig = px.box(df, 
             x='DATE',
             y='TAVG',
             template="plotly_dark"
            )
fig.show()
```

```
fig = px.box(df, 
             y='TAVG',
             color='DATE',
             animation_frame='DATE',
             range_y=[-10,30],
             template="plotly_dark",
             color_discrete_sequence=px.colors.cyclical.mygbm)
fig.show()
```

```
fig = px.violin(df, 
                x='TAVG',
                color='DATE',
                violinmode='overlay',
                animation_frame='DATE',
                orientation='h',
                range_x=[-20,40],
                color_discrete_sequence=px.colors.cyclical.mygbm,
                template="plotly_dark")
fig.show()
```

```
# importing the data as described in the txt
T_data = pd.read_csv('/content/data.txt',sep='\t',comment='%',names=['id', 'Series_Number', 'DATE', 'TEMP', 'Uncertainty', 'Observations','Time'])
```

```
# we need to convert the shit way they used to format data
# June 2008 = 2008 + (6 - 0.5)/12 = 2008.458
[months,years]=np.modf(T_data.DATE)
years = years.astype(int)
months = np.round(months*12+0.5).astype(int)

T_data.DATE=pd.to_datetime(years.map(str),format='%Y')
```

```
# double groupy to id and DATE
Y_temps = T_data.groupby(['id','DATE']).mean()
Y_temps = Y_temps.reset_index()
# round temps to 2 decimals
Y_temps.TEMP = np.round(Y_temps.TEMP,2).astype(float)
```

```
# now we merge the 2 dataframes based on id column
df = pd.merge(station_gps_data,Y_temps,on='id')
df = df.sort_values('DATE')
df = df.reset_index()
```

```
#shall we plot on a map?
fig = px.scatter_geo(df[df.DATE.dt.year==2019],
                     lat='lat',
                     lon='lon',
                     color='TEMP',
                     projection="natural earth",
                     template="plotly_dark")
fig.show()
```

```
#why not 3d?
fig = px.scatter_geo(df[df.DATE.dt.year==2019],
                     lat='lat',
                     lon='lon',
                     color='TEMP',
                     projection="orthographic",
                     template="plotly_dark")
fig.show()
```

Let's see if there are differences between years

```
year_1 = df[df.DATE.dt.year==2018]
year_2 = df[df.DATE.dt.year==2017]
```

```
m = pd.merge(year_1,year_2,how='inner',on='id')
```

```
m['DIFF'] = m.TEMP_x-m.TEMP_y
```

```
#shall we plot on a map?
fig = px.scatter_geo(m,
                     lat='lat_x',
                     lon='lon_x',
                     color='DIFF',
                     projection="natural earth",
                     range_color=[-1,1],
                     template="plotly_dark")
fig.show()
```
